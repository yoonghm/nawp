# 4.3 <span class='mn'>string.Template</span>

-   `string.Template` and f-string provide powerful and efficient ways to create template engine.

## <span class='mn'>string.Template</span>

-   `string.Template` provides string substitution.


-   Numerical values must be converted to string before they could be used


-   We could also use `jinja2` library for more powerful string substitutions

<span class='ex'>Example: <span class='mn'>string.substitute()</span></span>

```python
from string import Template

s1 = Template('$city is the capital of $country.')
t1 = s1.substitute(city='Kuala Lumper', country='Malaysia')
print(t1)

cc = {'country':'Malaysia', 'city':'Kuala Lumper'}
print(s1.substitute(**cc))
```

<span class='ex'>Example: <span class='mn'>string.substitute()</span> with <span class='mn'>$</span> symbol</span>

```python
from string import Template

cost = 2.50
unit = 5
s1 = Template('${unit}unit(s) * $$${cost} = $$${money}')
print(s1.substitute(cost=str(cost), unit=str(unit), money='%.2f' % (cost*unit)))
```

<span class='ex'>Example: More complicated example</span>

```python
## Use Template to create mass email
from string import Template

template = \
'''
Dear $student,

  Your scholarship interview will be on $date in $venue.
  
  See you then.
  
Best regards
$teacher
'''

letter = Template(template)

data = [{'student':'David Tan'   , 'date':'2019-01-08', 'venue':'23-05-02', 'teacher': 'Yoong Hor Meng' },
        {'student':'Micheal Li'  , 'date':'2019-01-08', 'venue':'23-05-03', 'teacher': 'Yoong Hor Meng' },
        {'student':'Sanjib Singh', 'date':'2019-01-09', 'venue':'23-05-02', 'teacher': 'Yoong Hor Meng' },
        {'student':'Abu Mohammad', 'date':'2019-01-09', 'venue':'23-05-03', 'teacher': 'Yoong Hor Meng' },
       ]

for d in data:
  print(letter.substitute(**d))
  print('-.'*40)
```
